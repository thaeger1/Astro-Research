Aiming to recreate cooling tables in Figure 2. from Wiersma et al. 2009 and Figure 1. from Thompson et al. 2015

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
import h5py
f = h5py.File('z_0.000.hdf5', 'r')
print(list(f.keys()))
s = 25 # slice in table ()

dset = f['Total_Metals']

dsub = dset['Net_Cooling']
y_plt_m = dsub[()][()]

dsub = dset['Temperature_bins']
temps = dsub[()]

dset = f['Metal_free']
dsub = dset['Net_Cooling']
y_plt_mf = dsub[()][()][()]
print(y_plt_mf.shape)

y_plt_mf = y_plt_mf[1,::,]
# print(temp.shape)
# print(temp[0,25])

# dsub = dset['Temperature_bins']
# x_plt_mf = dsub[()]

fig, ax = plt.subplots(figsize=(5, 2.7), layout='constrained')
# n = 0.003 is the 20th density value in the table, we can take the temp slice at this index and plot it
total = y_plt_m[:,s] + y_plt_mf[:,s]
ax.loglog(temps, np.abs(y_plt_m[:,s]), label='Metals')
ax.loglog(temps, np.abs(y_plt_mf[:,s]), label='H+He')
ax.loglog(temps, np.abs(total), label='Total')
ax.legend()
ax.set_xlabel('T (K)')
ax.set_ylabel('Net Cooling (ergs $cm^3 s^{-1}$)')
```
